```python
a = 1.0         # from given equation 1.0 == 1.0 + 2**(-53) is True. That means 1.0 + 2**(-53) = 1.0
b = 2 ** (-53)  # I take the numbers to prove associativity 
c = 2 ** (-53)  # thr rule of associativity: (a + b) + c = a + (b + c)

print('The result of "(a + b) + c" with "float" is', (a + b) + c) 

print('The result of "a + (b + c)" with "float" is', a + (b + c)) 

# Because of rounding, float numbers can give different results depending on how you add them. So addition is not associative for floats.
if ((a + b) + c ==  a + (b + c)):
    print('Associativity works')
else:
    print('We found such a counterexample, addition is not associative for "float".') 
```

```python
x = 1.0 + 2.0 **(-27)     # the defenitions of x and y that were given
y = 1.0 + 2.0 **(-28) 

result = x**2 - y**2  # whats we need to find

result_new = (x - y) * (x + y) # another form of difference of squares, that works better with float

print(result)
print(result_new)

from fractions import Fraction            # importing the Fraction function from the fractions module
x_f = Fraction(1) + Fraction(1, 2**27)    # allows us to perform exact calculations (no floating-point errors)
y_f = Fraction(1) + Fraction(1, 2**28)

result_check = x_f**2 - y_f**2            # the exact result of difference of squares

print(
    '\nComparison of x² - y² in float with the correct result is', 
    result == result_check, '\nThe reason: it loses precision due to rounding.\n')

print(
    'Comparison of (x - y) · (x + y) in float with the correct result is', result_new == result_check, 
    '\nThe reason: it avoids subtracting two nearly equal numbers.' )
```

```python
n = -1022                     # starting exponent (smallest exponent for normal float numbers)
last_x = None                 
last_n = None                 # will store the last non-zero float value and the corresponding exponent

while True:                   # run until we reach the limit of float representation
    x = 2.0**n                
    if x == 0:                # if x becomes 0 stop the loop
        break                 
    last_x = x                # save the last non-zero value of x and corresponding exponent
    last_n = n + 1                
    n -= 1                    # decrease exponent to get smaller and smaller numbers

print("Smallest positive float from 2 with negative power:", last_x)      
print("Corresponding exponent n:", last_n)     
```
