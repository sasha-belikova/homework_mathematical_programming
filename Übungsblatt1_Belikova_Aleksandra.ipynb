```python
for k in range(6):
    z = k * (2 * pi) / 6
    print('Für cos(z) = cosh(jz) mit k =', k, 'gilt:', cos(z) == cosh(1j * z))
    print('Für sin(z) = sinh(jz) / j mit k =', k, 'gilt:', sin(z) == (cmath.sinh(1j * z))/1j)
    print('Für sin^2(z) = 1 - cos^2(z) mit k =', k, 'gilt:', sin(z) **2 == 1 - cos(z) **2 )
```

```python
for k in range(6):
    z = cmath.rect(1, k * (2 * pi) / 6)
    print('Für cos(z) = cosh(jz) mit k =', k, 'gilt:', cmath.cos(z) == cosh(1j * z))
    print('Für sin(z) = sinh(jz) / j mit k =', k, 'gilt:', cmath.sin(z) == (cmath.sinh(1j * z))/1j)
    print('Für sin^2(z) = 1 - cos^2(z) mit k =', k, 'gilt:', cmath.sin(z) **2 == 1 - cmath.cos(z) **2 )
```
