## Beispel 2

```python
class Vektor3d:                                                  # beginning of a class for 3D vectors
    def __init__(self, x, y, z):                                 # constructor that defines how vectors are represented
        self.x = x                                                        
        self.y = y
        self.z = z

    def __add__(self, other):                                               # function that adds a scalar to every component of the vector
        if isinstance(other, (float, int)):                                  # checks if the scalar is of type float or int
            return Vektor3d(self.x + other, self.y + other, self.z + other)
        else:                                                                     # if not, raises an error
            raise TypeError('Addition with illegal type.')

    def __radd__(self, other):                                   # function that allows scalar + vector addition (reverse order)
        if isinstance(other, (float, int)):
            return Vektor3d(other + self.x, other + self.y, other + self.z)
        else:
            raise TypeError('Addition with illegal type.')
            
    def __str__(self):                                     # function that defines how vectors are displayed as strings, so it is eseier to read
        return f'({self.x}, {self.y}, {self.z})'

v = Vektor3d(1, 4, 3)
add = 6
result = add + v

print(f'The addition of vector {v} with scalar {add} is {result}')
```

```python
class VektorZMXY(Vektor3d):                                  
    def __init__(self, x, y, z):                             
        self.x = x                                           # same to previous
        self.y = -x                                          
        self.z = 0    

    def __lt__(self, other):                                # compares the x-coordinate of the first vector to the x-coordinate of the second one
        return self.x < other.x
        
    def __le__ (self, other):
        return self.x <= other.x

    def __gt__ (self, other):
        return self.x > other.x

    def __ge__(self, other):
        return self.x >= other.x

    def __str__(self):                                # function that defines how vectors are displayed as strings, so it is eseier to read
        return f'({self.x}, {self.y}, {self.z})'

v1 = VektorZMXY(9, 7, 9)
v2 = VektorZMXY(9, 6, 1)

a = v1 < v2
b = v1 <= v2
c = v1 > v2
d = v1 >= v2

print(f'The comparison {v1} < {v2} is {a}')
print(f'The comparison {v1} < or = {v2} is {b}')
print(f'\nThe comparison {v1} > {v2} is {c}')
print(f'The comparison {v1} > or = {v2} is {d}')
```
